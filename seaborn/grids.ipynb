```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
s=sns.PairGrid(iris,hue='species')
s.map(plt.scatter)
```

```python
tips = sns.load_dataset('tips')
g= sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
s = sns.FacetGrid(data=tips,col='time',row='smoker')
s.map(plt.scatter,'total_bill','tip')
```
