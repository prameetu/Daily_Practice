```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_title('title')
axes.set_xlabel('x')
axes.set_ylabel('y')
```

```python
fig = plt.figure()

ax1= fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y,color='red')
ax2.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2= fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Z')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig,axes= plt.subplots(1,2)
```

```python
axes[0].plot(x,y,color='blue',lw=2,ls='--')
axes[1].plot(x,z,color='green',lw=2,ls='-.')

fig
```
